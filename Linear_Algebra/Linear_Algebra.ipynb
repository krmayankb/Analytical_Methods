
### 4. Commutativity

In three dimensions, there are three different kinds of **rotations**, one for rotation about each axis. These are defined as follows:

$$
R_x(\theta) = 
\begin{pmatrix}
1 & 0 & 0 \\
0 & \cos \theta  & \sin \theta  \\
0 & -\sin \theta & \cos \theta 
\end{pmatrix}.
$$

$$
R_y(\theta) = 
\begin{pmatrix}
\cos \theta  & 0 & \sin \theta \\
0 & 1 & 0 \\
-\sin \theta & 0 & \cos \theta 
\end{pmatrix}.
$$

$$
R_z(\theta) = 
\begin{pmatrix}
\cos \theta  & \sin \theta & 0 \\
-\sin \theta & \cos \theta & 0 \\
0 & 0 & 1
\end{pmatrix}.
$$

Are these matrices pairwise commutative in general? That is: is it the case that

$$
R_x(\theta_1)R_y(\theta_2) = R_y(\theta_2)R_x(\theta_1)
$$

and similarly for $x,z$ and $y,z$? Explain your answers.

```python
from IPython.display import display

t = sm.symbols("t")
exp1 = "sin (t)"
exp2 = "cos (x)"
exp1 = sm.sympify(exp1)
exp2 = sm.sympify(exp2)

RxTheta = sm.Matrix([[1, 0, 0],[0, exp2, exp1], [0, -exp1,exp2] ]) 
RyTheta = sm.Matrix([[exp2, 0, exp1],[0, 1, 0], [-exp1, 0, exp2] ]) 
RzTheta = sm.Matrix([[exp2, exp1, 0],[-exp1, exp2, 0], [0, 0, 1] ]) 

############################# Checking for Rx.Ry = Ry.Rx ##############
print("RxTheta * RyTheta = ")
display(RxTheta * RyTheta)

print("RyTheta * RxTheta = ")
display(RyTheta * RxTheta)

print("Therefore, RyTheta and RxTheta are not commutative")
print("###################################################")
############################# Checking for Rx.Rz = Rz.Rx ##############
print("\nRzTheta * RxTheta = ")
display(RzTheta * RxTheta)

print("RxTheta * RzTheta = ")
display(RyTheta * RxTheta)

print("Therefore, RxTheta and RzTheta are not commutative")
print("###################################################")
############################# Checking for Ry.Rz = Rz.Ry ##############
print("\nRyTheta * RzTheta = ")
display(RyTheta * RzTheta)

print("RzTheta * RyTheta = ")
display(RzTheta * RyTheta)

print("Therefore, RyTheta and RzTheta are not commutative.")
print("###################################################")

print("Therefore, Mulitplication of rotation matrices are not communitative.")
```

### 5. Orthogonal Matrices

The set of **reflection matrices** is defined by

$$
S(\theta) = 
\begin{pmatrix}
\cos 2\theta & \sin 2 \theta \\
\sin 2\theta & -\cos 2 \theta

$$

a) Show how $S(\frac{\pi}{4})$ transforms the vector $(1\;-1)^T$. 

b) Show that $S(\theta)$ does not change any vectors that happen to lie on the axis of reflection.

```python
# Question 5 (a)
import numpy as np
t = sm.symbols(" t ")
exp1 = "sin (2*t)"                             # saving sin(x) as a string
exp1 = sm.sympify(exp1)                        # converting string to sympy expression.

exp2 = "cos (2*t)"                             # saving cos(x) as a string
exp2 = sm.sympify(exp2)                        # converting string to to sympy expression

STheta = sm.Matrix([[exp2, exp1],[exp1, -exp2]])  #declaring matrix as given in the question. 

STheta.subs({t: np.pi/4})                      # assigning t = π/4 

Mat = sm.Matrix([1 ,-1])                       # Declaration of vector as per question

print("Original Matrix: ")
display(Mat)
out =  Mat.transpose() * STheta.subs({t: np.pi/4})  
print("\nAfter Reflection: ")
display(out.transpose())
```

Recall that for matrices $A$ and $B$  it is not necessarily the case that $AB = BA$ (i.e. that $A$ and $B$ commute). Show that

___
#### Question 9 a)
\begin{align}
e^{A}e^{B} & = \left(1 + A + \frac{A^2}{2!} + \frac{A^3}{3!} …\right)\left(1 + B + \frac{B^2}{2!} + \frac{B^3}{3!} … \right)\\
           & = \left(\sum \frac{A^{m}}{m!}\right)\left(\sum\frac{B^{n}}{n!}\right)\\
           & =\sum^{\infty}_{m=0}\sum^{\infty}_{n=0}\frac{A^{m}B^{n}}{m!n!}\\\\
           & \text{if we consider, } l = m+n \text{ and } AB = BA\\
           & =\sum^{\infty}_{l=0}\sum^{l}_{m=0}\frac{A^{m}B^{l-m}}{m!(l-m)!}\\
           & =\sum^{\infty}_{l=0}\frac{1}{l!}\sum^{l}_{m=0}\frac{l!}{m!(l-m)!}A^{m}B^{l-m}\\\\
           & \text{if we refer binomial expansion, }\\
           & =\sum^{\infty}_{l=0}\frac{(A+B)^{l}}{l!}\\
           & = e^{A+B}
\end{align}

similarly, we can calculate for $e^{B}e^{A}$
\begin{align}
e^{B}e^{A} & = \left(1 + B + \frac{B^2}{2!} + \frac{B^3}{3!} … \right)\left(1 + A + \frac{A^2}{2!} + \frac{A^3}{3!} …\right)\\
           & = \left(\sum\frac{B^{n}}{n!}\right)\left(\sum \frac{A^{m}}{m!}\right)\\
           & =\sum^{\infty}_{n=0}\sum^{\infty}_{m=0} \frac{B^{n}A^{m} }{n!m!}\\\\
           & \text{if we consider, } l = m+n \text{ and } AB = BA\\
           & =\sum^{\infty}_{l=0}\sum^{l}_{n=0}\frac{B^{n}A^{l-n}}{n!(l-n)!}\\
           & =\sum^{\infty}_{l=0}\frac{1}{l!}\sum^{l}_{n=0}\frac{l!}{n!(l-n)!}B^{n}A^{l-n}\\\\
           & \text{if we refer binomial expansion, }\\

           & =\sum^{\infty}_{l=0}\frac{(A+B)^{l}}{l!}\\
