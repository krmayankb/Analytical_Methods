#### Answer to Question 2) a)
$\lim {\frac {3n+1}{2n+5}} = \frac{3}{2}$

To prove this, if for every $\epsilon > 0$, there is an N > 0, such that if n > N then $\vert T_n - L \vert < \epsilon $ 

To find N, Let us start with, \
$$\vert T_n - L \vert < \epsilon $$ 

$\Rightarrow$ $$ \left\vert \frac {3n+1}{2n+5} - \frac{3}{2}\right \vert  < \epsilon $$\
$\Rightarrow$ $$ \left\vert \frac{(6n+15) - (6n+2)}{4n+10} \right\vert < \epsilon $$\
$\Rightarrow$ $$ \frac{13}{\vert4n + 10\vert}  < \epsilon $$

as,$$ 4n < 4n + 10 \Rightarrow \frac{13}{4n} > \frac{13}{4n + 10}$$   

$\Rightarrow$ $$ \frac{13}{\vert 4 \epsilon\vert } < n$$   

$\Rightarrow$ $$ \frac {13}{4\epsilon} < N$$
$\Rightarrow$ $$ \frac {1}{N} < \frac{4\epsilon}{13} ... eq(1)$$ 

Proof: Now, we have $\epsilon > 0 $ and we choose $ N > \frac {13}{4\epsilon}$,

$\vert T_n - L \vert $ can be approximated to, \
$$\vert T_n - L \vert  = \left| \frac {3n+1}{2n+5} - \frac{3}{2} \right| $$ \
$$ =  \frac{13}{\vert4n + 10\vert}   $$\
$$ =  \frac{13}{\vert4n + 10\vert} < \frac{13}{\vert4n\vert}  $$
Since, $$n > N > \frac{13}{\vert 4\epsilon \vert}$$
$\Rightarrow$ $$ \epsilon > \frac {13} {4n} $$
$\Rightarrow$ $$  n > \frac {13} {4\epsilon} $$
Thus, $$ \left|\frac {3n+1}{2n+5} - \frac{3}{2} \right| \leq \frac {13} {4n} < \epsilon $$

Here, for N = $\frac{13}{4\epsilon} $ will be highest n for a given $\epsilon$. if we are reducing the value of $\epsilon$ further, we will get an n > N. This "n" will become updated N if we are doing more iterations. Apart from the above proof, if we look closely on the graph (ploted below) of $\vert T_n - L\vert$. We will find the $\epsilon$ converges to "0" but there is always a $\epsilon > 0 $ which will be available.

#### Question 2) b) 
$\lim \frac{2n}{n+2} = 2$

To prove this, if for every $\epsilon > 0 $, if there is an $ N > 0 $ such that if $ n > N $ then $\vert T_n - L\vert < \epsilon $   

Let us start with, $$\vert T_n - L\vert < \epsilon $$\
$\Rightarrow$ $$\left\vert \frac{2n}{n+2} - 2   \right\vert < \epsilon $$\
$\Rightarrow$ $$\left\vert \frac {2n - (2n+4)}{n+2} \right\vert < \epsilon $$\
$\Rightarrow$ $$ \frac{4}{\left\vert n+2 \right\vert } < \epsilon \;\;...eq(1)$$

As, $$ if \; n < n+2 \;\; then, \frac{4}{n} > \frac {4}{n+2} \;\;...eq(2) $$  
$\therefore \;\;\; from\;\;eq(1)\;\;and\;\;eq(2),$ $$ \frac {4}{n} < \epsilon $$ 
$\Rightarrow$ $$ \frac {4}{\epsilon} < n $$
$\Rightarrow$ $$ \frac {4}{\epsilon} < N $$
$\Rightarrow$ $$ \frac {4}{N} < \epsilon $$

Proof: Now, we have $\epsilon > 0 $ and we choose $ N > \frac {4}{\epsilon}$,\
as $\left\vert T_n - L\right\vert$ can be approximated to, 
$$\left\vert T_n - L\right\vert = \frac{4}{ | n+2 |}$$\
$$\left\vert T_n - L\right\vert = \frac{4}{ | n+2 |} < \frac{4}{ | n |} $$\
Since, $n > N > \frac {4}{\epsilon}$\
$\Rightarrow$ $$ \frac{4}{n} < \epsilon$$ \
Thus, $$\left\vert T_n - L\right\vert \leq \frac{4}{n} < \epsilon $$


Here, for N = $\frac{4}{\epsilon} $ will be highest "n" for a given $\epsilon$. if we are reducing the value of $\epsilon$ further, we will get an n > N. This "n" will become updated N if we are doing more iterations.
Above proves that for n > N, there exists and $\epsilon$ such that $\left\vert T_n - L\right\vert < \epsilon $.

                                        & \leq & \left |(x_n - L) \right | + \left |(y_n - M ) \right | ...(1)

$\Rightarrow$ $$ \frac 1{\left\vert n \right\vert }  < \epsilon $$

$\Rightarrow$  if we select, $$ \frac 1{\epsilon} < N$$

$\left|X_n - L\right|$  can be approximated to,
$$ \left\vert\frac{1}{n} - 0 \right\vert = \frac{1}{n} $$\
$\Rightarrow $ $$ \frac{1}{n} < \epsilon $$

Since, we have $$ n > N > \frac 1{\epsilon} $$
$\Rightarrow$ $$ \frac{1}{\epsilon} < n  $$
This, $$ |X_n - L| \leq \frac{1}{n} < \epsilon$$

Here, N = $\frac {1}{\epsilon}$ will be the highest value for a given $\epsilon$. there always exits, n > N such that $|\frac 1{n} - 0 | < \epsilon $, if we are reducing the value of $\epsilon$ it will generate new "n" which will become value of N for next iteration.

### 8. Definition of Derivative

Use the definition of the derivative to find the derivatives of the following functions.

a) $f(x) = x^3$, for any $x$

b) $g(x) = 1/x$, for $x \neq 0$

c) $h(x) = \sqrt{x}$, for $x > 0$

d) $k(x) = x^{-1/2}$ for $x > 0$.

___
### Answer to Q8
Derivative of a function f(x) can be defined as below: 
$$
f'(c) = \lim_{x \rightarrow c} \frac{f(x) - f(c)}{x-c}. 
$$
___
#### 8) a)

$f(x) = x^3$, for any $x$
So, $$ f'(c) =\lim_{x \rightarrow c} \frac{x^3 - c^3}{x-c}$$\
$$ =   \frac{(x-c)(x^2 + c^2 + xc)} {x-c}$$\
as sign for the (x-c) in Numerator(N) is same (x-c) in Denominator(D), $$ =   (x^2 + c^2 + xc) $$
 
Putting x = c,$$ =   (c^2 + c^2 + c*c) $$\
$$\therefore f'(c) = 3c^2 $$\
$$\Rightarrow f'(x) = 3x^2 $$
___
#### 8) b)

$g(x) = 1/x$, for $x \neq 0$
So, $$ g'(c) =\lim_{x \rightarrow c} \frac{\frac 1{x} - \frac1{c}}{x-c}$$\
$$ =   \frac{(c-x)} {(x-c) (cx)}$$\
$$ =   -\frac{(x-c)} {(x-c) (cx)}$$\
if we make the sign for the (x-c) in Numerator(N) and Denominator(D) same, $$ =   -\frac{1}{cx} $$
 
Putting x = c,$$ =   - \frac1{c*c} $$\
$$\therefore g'(c) =  - \frac1{c^2} $$\
$$\Rightarrow g'(x) =  - \frac1{x^2}$$
___
#### 8) c)
$h(x) = \sqrt{x}$, for $x > 0$
So, $$ h'(c) =\lim_{x \rightarrow c} \frac{ \sqrt{x} - \sqrt{c}}{x-c}$$\
Multiplying N and D with conjugate of N, we get: $$ = \frac{ (\sqrt{x} - \sqrt{c})(\sqrt{x} + \sqrt{c})}{(x-c)(\sqrt{x} + \sqrt{c})} $$\
$$ = \frac{ x-c }{(x-c)(\sqrt{x} + \sqrt{c})} $$\
$$ = \frac{1}{\sqrt{x} + \sqrt{c}} $$\
Putting x = c,$$ =   \frac{1}{\sqrt{c} + \sqrt{c}} $$\
$$\therefore f'(c) =  \frac1{2\sqrt{c}} $$\
$$\Rightarrow f'(x) =  \frac1{2\sqrt{x}}$$

___
#### 8) d)


$k(x) = x^{-1/2}$ for $x > 0$.
So, $$ k'(c) =\lim_{x \rightarrow c} \frac{x^{\frac {-1}{2}} -  c^{\frac {-1}{2}}  }{ x-c } $$\
$$ = \frac{ \sqrt{c} - \sqrt{x}  }{ \sqrt{cx}(x-c) } $$

Multiplying N and D with conjugate of N, we get: $$ = \frac{ (\sqrt{c} - \sqrt{x})(\sqrt{c} + \sqrt{x})}{\sqrt{cx}\;(x-c)(\sqrt{c} + \sqrt{x})} $$\
$$ = - \frac{ x-c }{(x-c)(\sqrt{cx}) (\sqrt{x} + \sqrt{c})} $$\
$$ = - \frac{1}{(\sqrt{x} + \sqrt{c})(\sqrt{cx}) } $$\
Putting x = c,$$ =   -\frac{1}{2\sqrt{c}  (\sqrt{c^2})} $$\
$$\therefore f'(c) =  -\frac 1{2  c^{\frac{3}{2}}} $$\
$$\therefore f'(x) =  -\frac 1{2  x^{\frac{3}{2}}} $$

### 9. The Chain Rule

Use the chain rule to find these deriviatives. Verify your results with `sympy`.

a) $f(x) = \frac{1}{1+x^2}$

b) $h(x) = (\sin x^k)^m$ for $m,k \in \mathbb{N}$

___
#### Answer to Question 9
a) $f(x) = \frac{1}{1+x^2}$\
$f(x)$ can be represented as $h(k(x))$, where $ h(x) = \frac 1{x} $ and $ k(x) = 1 + x^2 $\
$\Rightarrow $ $$h'(x) = -\frac{1}{x^2} \;\;\& \;\;k'(x) = 2x $$
$\Rightarrow $ $$ f'(x) = h'(k(x)) . k'(x) $$\
$$ = - \frac{1}{(1+x^2)^2} . 2x   $$\
$$ = -\frac{2x}{(1+x^2)^2} $$


___
b) $h(x) = (\sin x^k)^m$ for $m,k \in \mathbb{N}$\
$h(x)$ can be represented as $f(g(l(x)))$, where $ f(x) = x^m  $ and $ g(x) = sin x $ and $ l(x) = x^k $\
Here, $$h'(x) = f'(g(l(x))). g'(l(x)). l'(x)$$\
also, $f'(x) = m.x^{m-1}$ & $g'(x) = cos(x)$ & $l'(x) = k.x^{k-1}$

$$h'(k) = m.sin^{m-1}(x^k) . \frac{d}{dx} sin(x^k) $$\
$$h'(k) = m.sin^{m-1}(x^k) . cos(x^k). \frac{d}{dx} (x^k) $$\
$$h'(k) = m.sin^{m-1}(x^k) . cos(x^k). \frac{d}{dx}.k. (x^{k-1}) $$

Rearranging...
$$h'(k) = k.m.x^{k-1}. sin^{m-1}(x^k).cos(x^k) $$\
$$h'(k) = \frac {k.m.x^{k}.sin^{m}(x^k).cos(x^k)}{x.sin(x^k)} $$\
in above expression "." are mentioned only for ease of reading.
